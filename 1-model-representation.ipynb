```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x is the size of house in 1000 sq feet
# y is the price of house per 1000s of dollar

x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train {x_train}")
print(f"y_train {y_train}")
```

```python
# m is the number of training examples

m = x_train.shape

print(f"No. of training examples = {m}")
```

```python
len(x_train)
```

```python
i = 0

x_i = x_train[i]

y_i = y_train[i]

print(f"(x^{i}, y^{i}) = ({x_i}, {y_i})")
```

```python
plt.scatter(x_train, y_train, marker="x", c="r")
plt.title("House Prices")
plt.xlabel("Size of Houses per 1000 square feet")
plt.ylabel("Price of Houses per 1000s of dollar")
plt.show()
```

```python
w = 200
b = 100

print(f"W coefficient = {w}")
print(f"B coefficient = {b}")
```

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_w_b = np.zeros(m)
    
    for i in range(m):
        f_w_b[i] = w * x[i] + b
        
    return f_w_b
```

```python
temp_f_w_b = compute_model_output(x_train, w, b,)

plt.plot(x_train, temp_f_w_b, c='b',
            label="Our Prediction")

plt.scatter(x_train, y_train, c="r", marker="x",
            label="Actual Value")
plt.title("House Prices")
plt.xlabel("Size of Houses per 1000 square feet")
plt.ylabel("Price of Houses per 1000s of dollar")
plt.legend()
plt.show()
```
