```python
def z_score_normalize_feature(X):

    mu = np.mean(X, axis=0)
    sigma = np.std(X, axis=0)

    X_norm = (X - mu) / sigma

    return (X_norm, mu, sigma)
```

```python
mu = np.mean(X_train, axis=0)
sigma = np.std(X_train, axis=0)
X_mean = X_train - mu
X_norm = X_mean / sigma

fig, ax = plt.subplots(1, 3, figsize=(12,3))
ax[0].scatter(X_train[:,0], X_train[:,3])
ax[0].set_xlabel(X_features[0])
ax[0].set_ylabel(X_features[3])
ax[0].set_title("unnormalized")
ax[0].axis('equal')

ax[1].scatter(X_mean[:,0], X_mean[:,3])
ax[1].set_xlabel(X_features[0])
# ax[1].set_ylabel(X_features[3])
ax[1].set_title(r"X-$\mu$")
ax[1].axis("equal")

ax[2].scatter(X_norm[:,0], X_norm[:,3])
ax[2].set_xlabel(X_features[0])
# ax[2].set_ylabel(X_features[3])
ax[2].set_title(r"Z-score Normalized")
ax[2].axis("equal")

plt.show()
```

```python
X_norm, X_mu, X_sigma = z_score_normalize_feature(X_train)

print(f"X_Mu = {X_mu}, \n X_Sigma = {X_sigma}")
print(f"Peak to Peak range by column in raw            X:{np.ptp(X_train,axis=0)}")
print(f"Peak to Peak range by column in Normalized     X:{np.ptp(X_norm,axis=0)}")
```

```python
fig, ax = plt.subplots(1, 4, figsize=(12,3))

for i in range(len(ax)):
    norm_plot(ax[i], X_train[:,i])
    ax[i].set_xlabel(X_features[i])
ax[0].set_ylabel('count')
fig.suptitle("Distribution of feature before Normalization")
plt.show()

fig, ax = plt.subplots(1, 4, figsize=(12,3))

for i in range(len(ax)):
    norm_plot(ax[i], X_norm[:,i])
    ax[i].set_xlabel(X_features[i])
ax[0].set_ylabel('count')
fig.suptitle("Distribution of feature after Normalization")
plt.show()
```

```python
w_norm, b_norm, hist = run_gradient_descent(
    X_norm,
    y_train,
    iterations=1000,
    alpha=1e-1)
```

```python
plot_cost_i_w(X_norm, y_train, hist)
```

```python
m = X_norm.shape[0]
yp = np.zeros(m)

for i in range(m):
    yp[i] = np.dot(X_norm[i],w_norm) + b_norm

fig, ax = plt.subplots(1, 4, figsize=(15,5))

for i in range(len(ax)):

    ax[i].scatter(X_train[:,i], y_train, label="Target")
    ax[i].set_xlabel(X_features[i])
    ax[i].scatter(X_train[:,i], yp, color=dlc["dlorange"], label="Predicted")

ax[0].set_ylabel("Price")
ax[0].legend(loc=0)
fig.suptitle("Target vs Prediction using Z-score normalized model")
plt.show()
```
